# Random Walk

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1D Random Walk

```python
N=1000
a=np.random.randint(0,2,N)
y=np.cumsum(2*a-1)
x=np.arange(0,N,1)
plt.plot(x,y,'-')
plt.xlabel('Number of steps taken')
plt.ylabel('Displacement from origin')
```

## 2D Random Walk

```python
N=1000
a1=np.random.randint(0,2,N)
a2=np.random.randint(0,2,N)
y1=np.cumsum(2*a1-1)
y2=np.cumsum(2*a2-1)
plt.plot(y1,y2)
```

## Does random walk makes a gaussian?

## 1D gaussian

```python
final_pos=[]
for i in range(0,10000):
    N=500
    a=np.random.randint(0,2,N)
    final_pos.append(np.sum(2*a-1))
# fixed bin size
#bins = np.arange(-150, 150, 1.5) # fixed bin size

plt.hist(final_pos, bins=80)
plt.title('Number of counts')
plt.xlabel('Position')
plt.ylabel('count')

plt.show()
```

## 2D Gaussian

```python
final_posx=[]
final_posy=[]
for i in range(0,100000):
    N=5000
    ax=np.random.randint(0,2,N)
    ay=np.random.randint(0,2,N)
    final_posx.append(np.sum(2*ax-1))
    final_posy.append(np.sum(2*ay-1))
plt.hist2d(final_posx, final_posy, bins=(50,50),cmap=plt.cm.Reds)
```
